# Mean Filtering


```python
def mean_filter(image):
    copy_image = image.copy().astype(float)
    m,n = copy_image.shape
    output = np.empty_like(copy_image)
    
    for row in range (1, m-1):
        for col in range (1, n-1):
            total = copy_image[row-1, col-1] + copy_image[row-1, col] +copy_image[row-1, col+1] + \
                    copy_image[row, col-1] + copy_image[row, col] +copy_image[row, col+1] + \
                    copy_image[row+1, col-1] + copy_image[row+1, col] +copy_image[row+1, col+1]
            output[row, col] = total/9
    return output
```

```python
meanFiltered_image1 = mean_filter (copyCitra1)
meanFiltered_image2 = mean_filter (copyCitra2)

fig, axs = plt.subplots(2,2, figsize = (10,10))
ax = axs.ravel()

ax[0].imshow(copyCitra1, cmap ='gray')
ax[0].set_title('input gambar asli Citra 1')

ax[1].imshow(meanFiltered_image1, cmap ='gray')
ax[1].set_title('output mean filtered  Citra 1')

ax[2].imshow(copyCitra2, cmap ='gray')
ax[2].set_title('input gambar asli citra 2')

ax[3].imshow(meanFiltered_image2, cmap ='gray')
ax[3].set_title('output mean filtered  Citra 2')

plt.tight_layout()
```
